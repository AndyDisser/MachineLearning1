# Exercise 1: Min-Max normalization

### scale between $[a,b]$

$x^\prime = a + \frac{(x-min(x))*(b-a)}{max(x)-min(x)}$

```python
def min_max_scale(data, min_range=0, max_range=1):
    """the valuel of data gets changed in the function"""
    new_data = data
    min_value = np.amin(new_data)
    max_value = np.amax(new_data)
    print(f"The minimum value is: {min_value}")
    print(f"The maximum value is: {max_value}")
    new_data -= min_value
    new_data *= (max_range - min_range)
    new_data = np.divide(new_data, (max_value - min_value))
    
    return new_data
```

```python
feature_scaled_iris = np.zeros((150,4))
for i in range(len(iris_data)):
    feature_scaled_iris[i,:] = min_max_scale(iris_data[i,:])
    
pd.DataFrame(feature_scaled_iris, columns=["Sepal length","Sepal width","Petal length","Petal width"])
```

# b) scaling the entire dataset

```python
scaled_all_data = min_max_scale(iris_data)
pd.DataFrame(scaled_all_data, columns=["Sepal length","Sepal width","Petal length","Petal width"])
```
