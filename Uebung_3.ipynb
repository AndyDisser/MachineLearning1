```python
def gaussian(data, mue, sigma):
    """
    Calculate the probability of a multivariate gaussian.
    
    Parameters
    ----------
       - x: feature
       - mue: array of k mues
       - sigma: covariance matrix
       
    Returns
    -------
       - pobability for a point x in the given normal distribution N(mu,sigma)
    """
    scaling_factor = np.divide(1,(np.sqrt((2* np.pi)**len(mue) * np.linalg.det(sigma))))
    e_function = np.sum(-0.5*np.matmul((data-mue),np.linalg.inv(sigma))* (data-mue), axis=1)

    prob = scaling_factor * e_function
    return prob
```
